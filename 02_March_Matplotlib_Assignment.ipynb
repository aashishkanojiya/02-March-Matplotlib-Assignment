1.Line Plot: A line plot is a graph that displays data as a series of points connected by straight lines. This type of plot is useful for visualizing trends and changes over time.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Answer:-A scatter plot is a type of plot or mathematical diagram that uses Cartesian coordinates to display values for typically two variables for a set of data. Each point on the scatter plot represents one observation from the dataset, with the position of the point determined by the values of the two variables. Scatter plots are useful for visualizing the relationship between two continuous variables and for identifying patterns, trends, and potential correlations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X values')
plt.ylabel('Y values')

# Display the plot
plt.show()
```

Use the following data:

import numpy as np

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer:-The subplot() function in Matplotlib is used to create multiple plots within a single figure. This is particularly useful for comparing different datasets side-by-side or creating complex visualizations where multiple plots need to be presented together. It organizes plots in a grid-like structure, making them easy to compare and analyze.

Creating Four Line Plots Using subplot()

Here’s how you can use the subplot() function to create four line plots with the given data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the four line plots
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
axs[0, 0].plot(x, y1, 'r-')
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('X axis')
axs[0, 0].set_ylabel('Y axis')

# Second subplot
axs[0, 1].plot(x, y2, 'g-')
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('X axis')
axs[0, 1].set_ylabel('Y axis')

# Third subplot
axs[1, 0].plot(x, y3, 'b-')
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('X axis')
axs[1, 0].set_ylabel('Y axis')

# Fourth subplot
axs[1, 1].plot(x, y4, 'y-')
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('X axis')
axs[1, 1].set_ylabel('Y axis')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Answer:- A bar plot, or bar chart, is a type of chart that presents categorical data with rectangular bars. Each bar's height or length represents the value of the corresponding category. Bar plots are useful for comparing different categories of data and understanding their distribution or relationships.

Here are some specific reasons why bar plots are used:

1.Comparing quantities : Bar plots are used to compare the quantities of different categories, such as sales figures, revenue, or counts of items. By visually comparing the heights or lengths of the bars, it's easy to see which category has the largest or smallest value.

2.Tracking changes over time : Bar plots can also be used to track changes over time within a single category. For example, you might create a bar plot showing the revenue of a single department for each month of the year. By comparing the heights or lengths of the bars for each month, you can see how the revenue changes over time.

3.Showing relationships : Bar plots can also be used to show relationships between categories. For example, you might create a stacked bar plot showing the sales figures for different products in different regions. By stacking the bars on top of each other, you can see how the sales figures for each product compare across different regions.

4.Communicating results : Bar plots are a simple and effective way to communicate results to a non-technical audience. By presenting data in a visually appealing format, bar plots can help to convey the main points of your analysis quickly and clearly.

Plotting a Bar Plot and a Horizontal Bar Plot
Using the given data, here’s how you can create a bar plot and a horizontal bar plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(10, 5))
plt.bar(company, profit, color='blue')
plt.title('Profit by Company')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(10, 5))
plt.barh(company, profit, color='green')
plt.title('Profit by Company (Horizontal)')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Answer:- A box plot, also known as a box-and-whisker plot, is a statistical graph used to display the distribution of a dataset. It provides a visual summary of multiple aspects of the data, including its central tendency, variability, and potential outliers. The box plot includes the following components:

1.Median: The line in the middle of the box, representing the median (the middle value).

2.Interquartile Range (IQR): The box itself spans from the first quartile (Q1) to the third quartile (Q3), representing the middle 50% of the data.

3.Whiskers: Lines extending from the box to the smallest and largest values within 1.5 * IQR from the quartiles.

4.Outliers: Data points beyond the whiskers are considered outliers and are often plotted as individual points.

Box plots are used for several reasons:

1.Comparison: They allow for easy comparison of distributions between different datasets.

2.Visualization: They provide a clear and concise visual summary of the data's spread and central tendency.

3.Outliers: They help identify outliers and understand the variability within the data.

Plotting a Box Plot

Using the given data, here’s how you can create a box plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure
plt.figure(figsize=(8, 6))

# Create a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot of Two Datasets')
plt.xlabel('Dataset')
plt.ylabel('Values')

# Display the plot
plt.show()
```
